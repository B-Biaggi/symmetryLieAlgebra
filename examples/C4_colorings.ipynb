```python
is_graph_binomial = []
for same in colorings:
    flist = create_function_graphical_model(G,same[0]+same[1])
    #compute the symmetry Lie algebra
    symLie = symmetry_lie_algebra_random(flist)
    #decide whether the vanishing ideal is binomial
    binomial = test_binomiality_random(flist, symLie)
    is_graph_binomial.append([same[0]+same[1],binomial, len(symLie)])
```

```python
binomial_graph =  [res for res in is_graph_binomial if res[1] ==True ]
binomial_graph
```

```python
nonbinomial_graph =  [res for res in is_graph_binomial if res[1] ==False ]
```

```python
minim = nonbinomial_graph[0][2]
for res in nonbinomial_graph:
    if res[2] < minim:
        minim = res[2]
minim
```
