```python
# there are two different stages of leaves
stages = [1,2]
# for both stages, there are two branches
stages_children = {1:2, 2:2}

#create all the trees of depth 3
trees = trees_up_to_symmetry(3, stages_children , stages, True)
#there are so many of those
len(trees)
```

```python
# in fact, 15 of them are one-staged:
at_least_many_times = 1
trees_2 = keep_only_trees_with_all_stages_at_least(trees, stages, at_least_many_times)
len(trees_2)
```

```python
total = len(trees)
compute_nonbinomial_binary = []
for i in range(1,total):
    tree = trees[i]
    #create the parametrization
    flist  = tree_polynomials_hom(tree, stages_children)
    #compute the symmetry Lie algebra
    symLie = symmetry_lie_algebra_random(flist)
    #decide whether the vanishing ideal is binomial
    binomial = test_binomiality_random(flist, symLie)
    if binomial == False:
        compute_nonbinomial_binary.append([tree,binomial, len(symLie)])
```
