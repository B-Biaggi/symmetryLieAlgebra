Compute how many of them are not binomial and the minimal and maximal dimension of the symmetrie Lie algebra

```python
minim = nonbinomial_exp_3[0][2]
for res in nonbinomial_exp_3:
    if res[2] < minim:
        minim = res[2]
minim
```

```python
maxim = nonbinomial_exp_3[0][2]
for res in nonbinomial_exp_3:
    if res[2] > maxim:
        maxim = res[2]
maxim
```
